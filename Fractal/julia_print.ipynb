```python
import numpy as np
import fractal
from PIL import Image 
from matplotlib import cm
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
params = {
    'figure.figsize': (11,28),
}
plt.rcParams.update(params)
```

```python
width, height = 1920, 1080
xmin, xmax, ymin, ymax = -2.0, 2.0, -2.0, 2.0
maxiter = 512

c = 0.285 + 0.01j

julia = fractal.julia_set(xmin, xmax, ymin, ymax, width, height, maxiter,\
                               c=c, method='cuda')
```

```python
im = Image.fromarray(np.uint8(cm.gist_earth(julia)*255))
```

```python
plt.axis('off')
plt.imshow(im)
```

```python
im.save('julia_{}_{}.tif'.format(c.real, c.imag))
```
