```python
import numpy as np
import fractal
from PIL import Image 
from matplotlib import cm
import matplotlib.pyplot as plt
plt.style.use('dark_background')
```

```python
params = {
    'figure.figsize': (11,28),
}
plt.rcParams.update(params)
```

```python
width, height = 1920, 1080
xmin, xmax, ymin, ymax = -2.0,0.5,-1.25,1.25
maxiter = 512

mandelbrot = fractal.mandelbrot_set(xmin, xmax, ymin, ymax, width, height, maxiter, method='parallel')
```

```python
im = Image.fromarray(np.uint8(cm.gist_earth(mandelbrot)*255))
```

```python
plt.axis('off')
plt.imshow(im)
```

```python
im.save('mandelbrot.tif')
```
